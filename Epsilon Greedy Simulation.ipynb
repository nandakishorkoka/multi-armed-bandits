# Epsilon Greedy Simulation

```python
from EpsilonGreedyBandit import EpsilonGreedyBandit
from StaticEnvironment import StaticEnvironment
import seaborn as sns
import numpy as np
import random
```

```python
k = 10
rewards = [random.random() for i in range(k)]
print (rewards)
```

```python
egb = EpsilonGreedyBandit(k, "zero", 0.1)
env = StaticEnvironment(k,rewards)

average_rewards = []

for i in range(2000): 
    arm = egb.pull() 
    reward = env.evaluate(arm)
    egb.update(arm, reward)
    average_rewards.append(egb.average_reward())
    
print (egb)

sns.lineplot(data=np.array(average_rewards[:1500])) 
```
